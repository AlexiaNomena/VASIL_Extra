```python
output = widgets.Output()
AB = ["A", "B", "C", "D1", "D2", "E3", "E12", "F1", "F2", "F3", "NTD"]
from scripts2.Cross_neutralization.Compute_FR_major import Compute_FR_major
from scripts2.plotting.Cross_maps import Cross_maps
from scripts2.file_formats import Generate_txt
from scripts2.Expected_immunized.PNeut import PNeut
from scripts2.plotting.P_neut_groups import P_neut_groups

button_send = widgets.Button(
                description='Compute Results',
                tooltip='Send',
                style={'description_width': 'initial'}
            )

output = widgets.Output()

import numpy as np
import base64
def get_download_button(File):
    #https://stackoverflow.com/questions/61708701/how-to-download-a-file-using-ipywidget-button
    res = 'computed results'

    #FILE
    filename = File
    b64 = base64.b64encode(res.encode())
    payload = b64.decode()
    
    #BUTTONS
    html_buttons = '''<html>
    <head>
    <meta name="viewport" content="width=device-width, initial-scale=1">
    </head>
    <body>
    <a download="{filename}" href="data:text/csv;base64,{payload}" download>
    <button class="p-Widget jupyter-widgets jupyter-button widget-button mod-warning">Download File</button>
    </a>
    </body>
    </html>
    '''
    html_button = html_buttons.format(payload=payload,filename=filename)
    return html_button

def get_download_button2(File):
    #https://stackoverflow.com/questions/61708701/how-to-download-a-file-using-ipywidget-button
    res = 'computed results'

    #FILE
    filename = File
    b64 = base64.b64encode(res.encode())
    payload = b64.decode()
    
    #BUTTONS
    html_buttons = '''<html>
    <head>
    <meta name="viewport" content="width=device-width, initial-scale=1">
    </head>
    <body>
    <a download="{filename}" href="data:text/csv;base64,{payload}" download>
    <button class="p-Widget jupyter-widgets jupyter-button widget-button">Download SVG</button>
    </a>
    </body>
    </html>
    '''
    html_button = html_buttons.format(payload=payload,filename=filename)
    return html_button
    
def on_button_clicked(event):
    with output:
        clear_output()
        ### Generate required format
        lin1 = str(lineages_list[0].value)
        if lin1 == "":
            lin1 = "Lineage 1"
        lin2 = str(lineages_list[1].value)
        if lin2 == "":
            lin2 = "Lineage 2"
            
        mut1 = [str(positions[i])+str(mutations_list[0][i].value) if str(mutations_list[0][i].value)!="" else str("") for i in range(len(positions))]
        mut2 = [str(positions[i])+str(mutations_list[1][i].value) if str(mutations_list[1][i].value)!="" else str("") for i in range(len(positions))]
        Generate_txt(mut1, "Compare_variants/WebData/lineage1_mut.txt")
        Generate_txt(mut2, "Compare_variants/WebData/lineage2_mut.txt")
        print("Check mutations for %s:"%lin1, " ".join(list(np.array(mut1)[np.array(mut1)!=""])))
        print("Check mutations for %s:"%lin2, " ".join(list(np.array(mut2)[np.array(mut2)!=""])))
        print("")
        ### Compute cross
        Compute_FR_major(None, None, "data/epitope_data/dms_per_ab_per_site.csv", None, None, 2, [lin1, lin2], ["Compare_variants/WebData/lineage1_mut.txt", "Compare_variants/WebData/lineage2_mut.txt"],"Compare_variants/WebData/Cross_web.pck")
        ### Plot cross 
        how_WT = "not_add"
        if WT_Option.value == "YES":
            how_WT = "add"
        # generate plots
        Cross_maps("Compare_variants/WebData/Cross_web.pck", "Wuhan-Hu-1", None, "WebFigs", "True", how_WT)
        
        
        # view cross plots
        All_Cross = []
        for ab in AB:    
            file = open("WebFigs/major_Cross_React_AB_%s.png"%ab, "rb")
            image = file.read()
            image_cross = widgets.Image(
                value = image,
                format = "png",
                width = "300"
            )
            download = widgets.HTML(get_download_button2("WebFigs/major_Cross_React_AB_%s.svg"%ab))
            All_Cross.append(widgets.VBox([image_cross, download]))
        count = 0
        sc = 4
        Cross_Fig = []
        while count<len(All_Cross):
            Cross_Fig.append(widgets.HBox(All_Cross[count:count+sc]))
            count +=sc
                    
        BoxCross = widgets.VBox(Cross_Fig)

        ### Compute PNeut
        PNeut("data/Cross_with_delta_valitation.pck", "Compare_variants/WebData/Cross_web.pck", "data/epitope_data/dms_per_ab_per_site.csv", "data/Clinical_Delta_VE.xlsx", [lin1, lin2], [lin1, lin2], "Compare_variants/WebData")
        ### Plot PNeut
        P_neut_groups("data/PK_for_all_Epitopes.csv", "Compare_variants/WebData", "WebFigs", [lin1, lin2], ["orange", "#1f77b4"], lin1) 
        P_neut_groups("data/PK_for_all_Epitopes.csv", "Compare_variants/WebData", "WebFigs", [lin1, lin2], ["orange", "#1f77b4"], lin2) 
        ### View PNeut Plots
        file1 = open("WebFigs/P_Neut_groups_%s.png"%lin1, "rb")
        image1 = file1.read()
        image_pneut1 = widgets.Image(value = image1, format = "png", width = "700")
        download1 = widgets.HTML(get_download_button2("WebFigs/P_Neut_groups_%s.svg"%lin1))


        file2 = open("WebFigs/P_Neut_groups_%s.png"%lin2, "rb")
        image2 = file2.read()
        image_pneut2 = widgets.Image(value = image2, format = "png", width = "700")
        download2 = widgets.HTML(get_download_button2("WebFigs/P_Neut_groups_%s.svg"%lin2))

        PNeut_Figs = widgets.VBox([image_pneut1, download1, image_pneut2, download2])

        fig = widgets.HBox([BoxCross, PNeut_Figs])
        display(fig)
        
button_send.on_click(on_button_clicked)
Get_res = widgets.VBox([button_send, widgets.HTML(value="<h4>Can take between 2 and 10 min</h4>")])
vbox_result = widgets.VBox([Get_res, output])
```
